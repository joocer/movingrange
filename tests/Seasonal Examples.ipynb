```python
import numpy as np
import pandas as pd
import timeseries as ts
```

Open the dataset, build a list from the desired feature

```python
cve_df = pd.read_csv(r'..\data\sea_ice.csv')
series = cve_df['Arctic'].tolist()
```

predict the period of the pattern - this works best when there is a strong pattern

```python
ts.cycle_periods(series)
```

Decompose the series into it component parts:

- source
- trend (linear regression)
- seasonal
- detrended
- residual

plot() plots each, with a mean line, residual also plots a zero line.

```python
decomposed = ts.decompose(series, 12)
decomposed.plot()
```

other trend and summary options include

```python
from matplotlib import pyplot as plt

plt.figure(figsize=(20, 6))
# henderson moving average
plt.plot(ts.henderson(series,35))
pass
```

```python
from matplotlib import pyplot as plt

plt.figure(figsize=(20, 6))
# moving average
plt.plot(ts.moving_average(series,7))
pass
```
